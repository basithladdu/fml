```python
import seaborn as sns
import matplotlib.pyplot as plt

def graph(y):
    plt.hist(y, bins=5)

plt.figure(figsize=(10,10))

# Adding the subplot at the specified
# grid position
plt.subplot(221)
graph(dataset['sepallength'])

plt.subplot(222)
graph(dataset['sepalwidth'])

plt.subplot(223)
graph(dataset['petallength'])

plt.subplot(224)
graph(dataset['petalwidth'])

plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


fig, axes = plt.subplots(2, 2, figsize=(10,10))

axes[0,0].set_title("Sepal Length")
axes[0,0].hist(dataset['sepallength'], bins=7)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(dataset['sepalwidth'], bins=5);

axes[1,0].set_title("Petal Length")
axes[1,0].hist(dataset['petallength'], bins=6);

axes[1,1].set_title("Petal Width")
axes[1,1].hist(dataset['petalwidth'], bins=6);
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(dataset.corr(), cmap='coolwarm', annot=True)

#sns.heatmap(dataset.corr(method='pearson'), acmap='coolwarm', annot=True)
plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

def graph(y):
    sns.boxplot(x="class", y=y, data=dataset)

plt.figure(figsize=(10,10))

# Adding the subplot at the specified
# grid position
plt.subplot(221)
graph('sepallength')

plt.subplot(222)
graph('sepalwidth')

plt.subplot(223)
graph('petallength')

plt.subplot(224)
graph('petalwidth')

plt.show()
```
