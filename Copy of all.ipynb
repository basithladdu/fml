```python

# 8 tree

import pandas as pd 
import numpy as np 
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df3=pd.read_csv('sample.csv')
df3

```

```python
le = LabelEncoder()
for col in df3.columns:
    df3[col] = le.fit_transform(df3[col])
x=df3.drop(columns=['Country'])
y=df3['Country']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model= DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
plt.figure(figsize=(12, 6))
plot_tree(model, feature_names=x.columns, class_names=[str(cls) for cls in set(y)], filled=True)
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
cm = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}\n')
print('Confusion Matrix:\n', cm)
print('\nClassification Report:\n', class_report)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
matrix = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
matrix.plot()
plt.show()
```

```python



# 10 Implement K Means Clustering algorithm


import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score
```

```python
df=pd.read_csv('Clustering.csv')
x = df[["Age", "Annual_Income", "Spending_Score"]]

kmeans=KMeans(n_clusters=3,random_state=42)
df["cluster"]=kmeans.fit_predict(x)


iner = kmeans.inertia_
silhouette = silhouette_score(x,df["cluster"])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Annual_Income"], df["Spending_Score"], c=df["cluster"], cmap='viridis', alpha=0.6)
plt.xlabel("Annual Income")  # Setting x-axis label
plt.ylabel("Spending Score")  # Setting y-axis label
plt.title("K-Means Clustering - Customer Segmentation")  # Setting plot title
plt.colorbar(label="Cluster")  # Adding color bar to indicate clusters
plt.show()  # Displaying the plot
```

```python

print(f"Total Samples: {len(df)}")  
print(f"Cluster counts:\n{df['cluster'].value_counts()}") 
print(f"Inertia: {iner}") 
print(f"Silhouette Score: {silhouette}")
```
