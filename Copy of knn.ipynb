```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report , recall_score,f1_score,precision_score,confusion_matrix
```

```python
df=pd.read_csv('iris_csv.csv')


x=df.drop(columns=['class']).values
y=df['class'].values


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=KNeighborsClassifier()
model.fit(x_train,y_train)
model
```

```python
y_pred=model.predict(x_test)
```

```python
print("acc",accuracy_score(y_test,y_pred))
print("recall_score",recall_score(y_test,y_pred,average='macro'))
print("precision",precision_score(y_test,y_pred,average='macro'))
cd=confusion_matrix(y_test,y_pred)
print(cd)
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(cd ,annot=True ,fmt='d',cmap='Blues')
```

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import ConfusionMatrixDisplay,classification_report ,accuracy_score
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('sample.csv')
df.columns
```

```python
le=LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
x=df.drop(columns=['Country'])
y=df['Country']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model= DecisionTreeClassifier()
model.fit(x_train,y_train)


y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)*100

c=confusion_matrix(y_test,y_pred)
```

```python
plt.figure(figsize=(12,6))
plot_tree(model,feature_names=x.columns,filled=True)
```

```python
matrix =ConfusionMatrixDisplay(confusion_matrix=c,display_labels=model.classes_)
matrix.plot()
```

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import ConfusionMatrixDisplay,classification_report ,accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('linear.csv')
df=pd.DataFrame(df)

x=df.drop('y',axis=1)
y=df['y']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
model_1 = LinearRegression()
model_2 = RandomForestRegressor()
model_3 = SVR()
```

```python
model_1.fit(x_train,y_train)
model_2.fit(x_train,y_train)
model_3.fit(x_train,y_train)
```

```python
y_predict1=model_1.predict(x_test)
y_predict2=model_2.predict(x_test)
y_predict3=model_3.predict(x_test)
```

```python
pred_final = (y_predict1+y_predict2+y_predict3)/3.0
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, pred_final))
```
