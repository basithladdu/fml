```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv("linear.csv")
X = df[['X']].values
y = df[['y']].values

split_sizes = [0.2, 0.3, 0.4, 0.5]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

colors = [
    ('deepskyblue', 'crimson'),
    ('mediumseagreen', 'darkorange'),
    ('mediumpurple', 'saddlebrown'),
    ('hotpink', 'darkslategray')
]

for i in range(len(split_sizes)):
    split_size = split_sizes[i]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_size, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    ax = axes[i // 2, i % 2]
    actual_color, predicted_color = colors[i]
    ax.scatter(X_test, y_test, color=actual_color, label='Actual')
    ax.plot(X_test, y_pred, color=predicted_color, linewidth=2, label='Predicted')
    ax.set_title(f'Split: {split_size}')
    ax.set_xlabel('X')
    ax.set_ylabel('y')
    ax.legend()

plt.tight_layout()
plt.show()
```
